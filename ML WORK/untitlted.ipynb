```python
def calculate_angle(a,b,c):
    a = np.array(a)  #first
    b = np.array(b)  #mid
    c = np.array(c)  #end
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])   #calculating the angle between sholder elbow and wrist
    angle = np.abs(radians*180.0/np.pi)  #converting the angle from radians to degree

    if angle >180.0:  
        angle = 360-angle #becoz hand cant go more than 180* 

    return angle
```

```python
# we are testing our new camera feed
cap=cv2.VideoCapture(0)# it is used to enable camera and 0 is the slot that is assigned for the video capturing device

#curl counter Variable
counter = 0 
stage = None

with mp_pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5) as pose: # accuracy of our tracker is set (if we set more accuracy the detection will be more precise)
    while cap.isOpened():
        ret,frame = cap.read()
        
        #Recolor Image to RGB
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)# here we are coloring the frame
        image.flags.writeable= False #setting the frame whether to take it or not
        
        #Make detection
        results= pose.process(image) 
        
        #Recolor image to BGR  
        image.flags.writeable = True
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)#here it reverts the picture to its original format ie from RGB to BGR
        

        #Extract landmarks
        try:
            landmarks=results.pose_landmarks.landmark

            shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]   #it fetches the landmarks that are noted before and takes the coordinates
            elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]  
            wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]  
        
            angle = calculate_angle(shoulder, elbow, wrist)
            #visulaize
            cv2.putText(image, str(angle),tuple(np.multiply(elbow, [640,480]).astype(int)),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)  #used to visualize and render and rhen determininf where is the array postioning
        #curl counter logic
            if angle > 160:
                stage = "down" 
            if angle < 30 and stage =='down':
                stage = "up"
                counter +=1
                print(counter)

        except:
            pass
        
        #Render Detections
        #setup status box
        cv2.rectangle(image, (0,0), (255,73), (245,117,16), -1)
        
        
        #Rep Data
        cv2.putText(image, 'REPS', (15,12), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, str(counter), (10,60), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2, cv2.LINE_AA)
        
        #Stage Data
        cv2.putText(image, 'REPS', (65,12), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, stage, (60,60), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2, cv2.LINE_AA)

        #render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS, #it uses utilites of the data. This is used to collect after scanning the image continously
                                mp_drawing.DrawingSpec(color=(245,117,82), thickness=2, circle_radius=2),
                                mp_drawing.DrawingSpec(color=(246,67,232), thickness=2, circle_radius=2)
        )
        cv2.imshow('Mediapipe feed',image) # enables and shows the video capturing application

        if cv2.waitKey(10) &0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
